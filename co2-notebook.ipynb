```python


num_cols = df.select_dtypes(include="number").columns

df[num_cols] = df.groupby("country")[num_cols].transform(
    lambda g: g.interpolate(limit_direction="both")
)

df_clean = df.fillna(0, inplace=False)

df.to_csv("owid-co2-data-interpolated.csv", index=False)
df_clean.to_csv("owid-co2-data-cleaned.csv", index=False)
```
